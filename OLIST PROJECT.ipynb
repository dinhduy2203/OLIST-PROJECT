```python

import matplotlib.pyplot as plt
import pandas as pd
link_data_1 = 'https://drive.google.com/file/d/1A6mT7Gr6vHB9kV6iY7JbvTuqs8U1jtLZ/view?usp=sharing'
data_1 = 'https://drive.google.com/uc?export=download&id=' + link_data_1.split('/')[-2]
df_customers = pd.read_csv(data_1)
df_customers.head()
```

```python
df_customers.info()
```

```python
file = df_customers.to_csv("Olist_customers_cleaned.csv")
```

```python
# Counting the number of customers in each state
state_counts = df_customers['customer_state'].value_counts()

# Plotting the distribution of 'customer_state'
plt.figure(figsize=(15, 6))
ax = state_counts.plot(kind='bar')  # Displaying all states
plt.title('Number of Customers by State')
plt.xlabel('State')
plt.ylabel('Number of Customers')

# Adding the count above each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Converting 'order_approved_at' to datetime and extracting the quarter
df_orders['order_approved_quarter'] = pd.to_datetime(df_orders['order_approved_at']).dt.to_period('Q')

# Counting the number of orders per quarter
quarterly_counts = df_orders['order_approved_quarter'].value_counts().sort_index()

# Plotting the distribution of 'order_approved_quarter'
plt.figure(figsize=(15, 6))
ax = quarterly_counts.plot(kind='bar')
plt.title('Distribution of Order Approvals by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Orders Approved')

# Adding the count above each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Counting the distribution of 'order_status'
order_status_distribution = df_orders['order_status'].value_counts()

# Plotting the distribution of 'order_status'
plt.figure(figsize=(10, 6))
ax = order_status_distribution.plot(kind='bar')
plt.title('Distribution of Order Status')
plt.xlabel('Order Status')
plt.ylabel('Frequency')

# Adding count labels above each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.xticks(rotation=0)
plt.show()
```
